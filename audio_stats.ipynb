```python
import audio_stats
from IPython.display import clear_output
from ipyfilechooser import FileChooser

# Create and display a FileChooser widget
fc = FileChooser('.')
# Set a file filter pattern (uses https://docs.python.org/3/library/fnmatch.html)
fc.filter_pattern = ['*.wav', '*.mp3']
# Callback example

def show_stats(chooser):
    stats = audio_stats.AudioStats()
    stats.load(chooser.selected)
    stats.analyze()
    clear_output(wait=True)
    stats.print()

# Register callback function
fc.register_callback(show_stats)
display(fc)
```
